```python
punctuations = """`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ""" + string.punctuation

text_en_lower=text_en.lower()

text_en_clean_punct = "".join([x for x in text_en_lower if x not in punctuations])
word_tok_en=word_tokenize(text_en_clean_punct)
print(word_tok_en)
```

```python
model_en = Word2Vec([word_tok_en],min_count=1)
model_en.save("Word2cev_en.model")
for i in word_tok_en:
    vt = model_en.wv[i]
    print(" {}= {}\n\n".format(i,vt))
```

# tf idf

```python
sentences = sent_tokenize(text_en)
print(sentences)
total_documents = len(sentences)
```

```python
sentences_ar = sent_tokenize(text_ar)
print(sentences_ar)
total_documents_ar = len(sentences_ar)
```

```python
def _create_frequency_matrix(sentences):
    frequency_matrix = {}
    stopWords = set(stopwords.words("english"))
    ps = PorterStemmer()

    for sent in sentences:
        freq_table = {}
        words = word_tokenize(sent)
        for word in words:
            word = word.lower()
            word = ps.stem(word)
            if word in stopWords:
                continue

            if word in freq_table:
                freq_table[word] += 1
            else:
                freq_table[word] = 1

        frequency_matrix[sent[:15]] = freq_table

    return frequency_matrix
```

```python
frequency=_create_frequency_matrix(sentences)
print(frequency)
```

```python
frequency_ar=_create_frequency_matrix(sentences_ar)
print(frequency_ar)
```

```python
def _create_tf_matrix(freq_matrix):
    tf_matrix = {}

    for sent, f_table in freq_matrix.items():
        tf_table = {}

        count_words_in_sentence = len(f_table)
        for word, count in f_table.items():
            tf_table[word] = count / count_words_in_sentence

        tf_matrix[sent] = tf_table

    return tf_matrix
```

```python
tf=_create_tf_matrix(frequency)
print(tf)
```

```python
tf_ar=_create_tf_matrix(frequency_ar)
print(tf_ar)
```

```python
def _create_documents_per_words(freq_matrix):
    word_per_doc_table = {}

    for sent, f_table in freq_matrix.items():
        for word, count in f_table.items():
            if word in word_per_doc_table:
                word_per_doc_table[word] += 1
            else:
                word_per_doc_table[word] = 1

    return word_per_doc_table
```

```python
doc=_create_documents_per_words(frequency)
print(doc)
```

```python
doc_ar=_create_documents_per_words(frequency_ar)
print(doc_ar)
```

```python
def _create_idf_matrix(freq_matrix, count_doc_per_words, total_documents):
    idf_matrix = {}

    for sent, f_table in freq_matrix.items():
        idf_table = {}

        for word in f_table.keys():
            idf_table[word] = math.log10(total_documents / float(count_doc_per_words[word]))

        idf_matrix[sent] = idf_table

    return idf_matrix
```

```python
idf=_create_idf_matrix(frequency,doc,total_documents)
print(idf)
```

```python
idf_ar=_create_idf_matrix(frequency_ar,doc_ar,total_documents_ar)
print(idf_ar)
```

```python
def _create_tf_idf_matrix(tf_matrix, idf_matrix):
    tf_idf_matrix = {}

    for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), idf_matrix.items()):

        tf_idf_table = {}

        for (word1, value1), (word2, value2) in zip(f_table1.items(),
                                                    f_table2.items()):  # here, keys are the same in both the table
            tf_idf_table[word1] = float(value1 * value2)

        tf_idf_matrix[sent1] = tf_idf_table

    return tf_idf_matrix
```

```python
tf_idf=_create_tf_idf_matrix(tf,idf)
print(tf_idf)
```

```python
tf_idf_ar=_create_tf_idf_matrix(tf_ar,idf_ar)
print(tf_idf_ar)
```
